```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
plt.figure(figsize = (10,6))
plt.plot(data['Altitude'] , data['Temprature'],  marker='o', linestyle='-', linewidth = 1.5 , markersize = 1)
plt.title('temprature variation')
plt.xlabel('altitude')
plt.ylabel('temperature')
plt.grid(True)
plt.show
```

```python
plt.figure(figsize = (10,6))
plt.plot(data['Altitude'] , data['Pressure'],  marker='o', linestyle='-' , linewidth = 1.5 , markersize = 1)
plt.title('pressure variation')
plt.xlabel('altitude')
plt.ylabel('pressure')
plt.grid(True)
plt.show
```

```python
plt.figure(figsize = (12,6))
plt.plot(data['Altitude'] , data['Humidity'],  marker='o', linestyle='-',linewidth=1.5, markersize=3)
plt.title('humidity variation with altitude')
plt.xlabel('altitude')
plt.ylabel('humidity')
plt.grid(True)
plt.show
```

```python

# Set the style
sns.set(style='whitegrid')

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Altitude', y='AQI', data=data, marker='o', linewidth=2.5)

# Customize the title and labels
plt.title('AQI Variation Over altitude', fontsize=16)
plt.xlabel('altitude', fontsize=14)
plt.ylabel('AQI', fontsize=14)

# Highlight a specific point (e.g., the highest AQI)
highlight = data[data['AQI'] == data['AQI'].max()]
sns.scatterplot(x='Altitude', y='AQI', data=highlight, color='red', s=100)

# Show the grid
plt.grid(True)
```

```python

# Create the plot
plt.figure(figsize=(10,6))
plt.plot(data['Altitude'] , data['Smoke'],  marker='o', linestyle='-',linewidth=1.5, markersize=1)
# Customize the title and labels
plt.title('smoke Variation with Altitude', fontsize=16)
plt.xlabel('Altitude (m)', fontsize=14)
plt.ylabel('smoke', fontsize=14)


# Show the grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(data['Altitude'] , data['velocity'],  marker='o', linestyle='-',linewidth=1.5, markersize=1)
plt.title('velocity variation')
plt.xlabel('altitude')
plt.ylabel('velocity')
plt.grid(True)
plt.show
```
