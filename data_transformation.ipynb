```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample negatively skewed data
data = pd.Series([-20, -10, -5, -3, -2, -1, 0, 1, 2, 2, 3, 3, 4, 4, 5])

# Square transformation
squared_data = np.power(data, 2)

# Visualization
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True)
plt.title('Original Data')

plt.subplot(1, 2, 2)
sns.histplot(squared_data, kde=True)
plt.title('Squared Data')
plt.show()
```

```python
# Exponential transformation
exp_data = np.exp(data)

# Visualization
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data, kde=True)
plt.title('Original Data')

plt.subplot(1, 2, 2)
sns.histplot(exp_data, kde=True)
plt.title('Exponential Transformed Data')
plt.show()
```
